```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output
import time
import random

# Helper & Mock Functions Assuming Imports From Parts 1-3
def kmeans_clustering(grid, k=3, max_iter=100):
    """
    K-Means Unsupervised ML — divides open cells into K patrol zones.
    Uses Euclidean distance. Stops when centroids converge.
    Returns: (labels array, centroids, iterations)
    """
    open_cells = np.array([(r, c) for r in range(len(grid)) 
                            for c in range(len(grid[0])) if grid[r][c] == 1])
    
    # Initialize centroids randomly from open cells
    indices = np.random.choice(len(open_cells), k, replace=False)
    centroids = open_cells[indices].astype(float)
    
    labels = np.zeros(len(open_cells), dtype=int)
    
    for iteration in range(max_iter):
        # Assign step — find closest centroid for each cell
        dists = np.linalg.norm(open_cells[:, None] - centroids[None, :], axis=2)
        new_labels = np.argmin(dists, axis=1)
        
        # Update step — recompute centroid as mean of cluster
        new_centroids = np.array([
            open_cells[new_labels == ki].mean(axis=0) if (new_labels == ki).any() 
            else centroids[ki] 
            for ki in range(k)
        ])
        
        print(f"K-Means: Iteration {iteration+1} — Centroids: {new_centroids.tolist()}")
        
        if np.allclose(new_centroids, centroids):
            print(f"K-Means: Iteration {iteration+1} — Centroids converged — clustering complete")
            return new_labels, new_centroids, iteration+1
        
        centroids = new_centroids
        labels = new_labels
    
    return labels, centroids, max_iter

# --- Interactive Game UI Setup ---
# Note: For full interactive running in jupyter notebook, the UI needs true interactive backends which we implement via clear_output.
btn_new_dungeon = widgets.Button(description="⟳ New Dungeon")
btn_run_ai = widgets.Button(description="▶ RUN AI")
algo_selector = widgets.ToggleButtons(
    options=['BFS', 'DFS', 'A*', 'Hill Climbing', 'Minimax', 'CSP', 'K-Means'],
    description='Algorithm:',
)
speed_slider = widgets.FloatSlider(min=0.01, max=1.0, value=0.1, description='Speed')
stats_panel = widgets.Output()
log_panel = widgets.Output()

display(algo_selector, speed_slider, widgets.HBox([btn_new_dungeon, btn_run_ai]))
display(widgets.HBox([stats_panel, log_panel]))

def on_new_dungeon(b):
    with log_panel:
        print("Generated new dungeon! Please run your code segments from Parts 1-3 to support the respective algorithm handlers or link them appropriately.")
    
btn_new_dungeon.on_click(on_new_dungeon)
```
